````{admonition} Venstretiln칝rming
:class: tip
Strategien er som f칮lger:
1. Del opp intervallet $[a, b]$ i $n$ like store delintervaller. Dette gir $n$ rektangler!
2. Regn ut h칮yden til v칝rt rektangel ved 친 bruke funksjonsverdien $f(x)$ i venstre ende av hvert delintervall.
3. Summer opp arealene til alle rektanglene for 친 finne en tiln칝rming til arealet mellom grafen og $x$-aksen.

Metoden kalles for *venstretiln칝rming* fordi vi bruker venstre ende av hvert delintervall til 친 finne h칮yden til rektanglene.

```{figure} ./figs/areal_figurer.png
````

I denne oppgaven skal du l칝re om venstretiln칝rming og bruke dette til 친 finne en tiln칝rming til integralet

$$
\int\limits_{0}^{3} -4(x-3)(x^2 + 1) \, dx.
$$

En kode som bruker venstretiln칝rming kan se slik ut:

`````python
def f(x):
    return -4 * (x - 3) * (x**2 + 1)

a = 0                                       # Nedre integrasjonsgrense i intervallet [a, b]
b = 3                                       # 칒vre integrasjonsgrense i intervallet [a, b]
n = 3                                       # Antall delintervaller (antall rektangler)
bredde = (b - a) / n                        # Bredden p친 hvert rektangel

integral = 0                                # Oppretter en variabel som lagrer summen av arealene til rektanglene
for i in range(n):                          # G친r gjennom hvert rektangel
    x = a + i * bredde                      # Regner ut x-verdien til venstre side av rektangelet
    h칮yde = f(x)                            # Regner ut h칮yden til rektangelet
    areal = h칮yde * bredde                  # Regner ut arealet av rektangelet
    integral = integral + areal             # Legger til arealet av rektangelet til summen

print(f"{integral = :.2f}")                 # Skriver ut verdien med 2 desimaler
`````
