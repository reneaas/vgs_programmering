## Lengden av en graf

Tenk deg at vi har en funksjon $f(x)$ og vi 칮nsker 친 regne ut lengden til grafen over et intervall $[a, b]$. M친ten vi kan tiln칝rme lengden dens p친, er 친 dele opp intervallet i $n$ delintervaller slik at vi f친r $n + 1$ punkter $x_0, x_1, \ldots, x_{n-1}$ som er jevnt fordelt i intervallet. For hvert punkt $x_i$, kan vi regne ut tilsvarende funksjonsverdier $y_i = f(x_i)$ slik at vi f친r $n + 1$ $y$-verdier $y_0, y_1, \ldots, y_n$. Med denne oppbygningen kan vi skrive ned resultatet som vi baserer oss p친:

```{prf:theorem} Lengde av graf
:label: lengde_av_graf

La $f(x)$ v칝re en funksjon definert p친 et intervall $[a, b]$. La $x_0, x_1, \ldots, x_n$ v칝re $n + 1$ punkter som er jevnt fordelt i intervallet. Da kan vi lage oss $n$ rettvinklede trekanter med kateter

\begin{align*}
\Delta x & = (b - a) / n, \\
\\
\Delta y_i & = f(x_i + \Delta x) - f(x_i),
\end{align*}

der $x_i = a + i\Delta x$ for $i = 0, 1, 2, \ldots, n$. 

Trekantene har da hypotenuser med lengde

$$
\Delta s_i = \sqrt{\Delta x^2 + \Delta y_i^2},
$$

Lengden av grafen til $f(x)$ over intervallet $[a, b]$ blir da tiln칝rmet gitt ved

$$
L \approx \sum_{i=0}^{n - 1} \Delta s_i = \Delta s_0 + \Delta s_1 + \ldots + \Delta s_{n - 1},
$$
```

St칮rrelsen $\Delta s$ kalles for et **linjeelement** og er et konsept som dukker opp i fysikken i mange sammenhenger.

### Eksempel 1

La oss bruke algoritmen over til 친 regne ut lengden av grafen

$$
f(x) = x^2
$$

p친 intervallet $[1, 4]$ slik som i {numref}`lengde_graf`. Vi skal bruke $n = 100$ punkter i stedet for $n = 4$ som i figuren. 

Vi kan kode dette slik:

`````python
import numpy as np

def f(x):
    return x**2

# Definerer intervallet [a, b] og antall punkter n
a = 1
b = 4
n = 10
dx = (b - a) / n # Steglengde 풊x

lengde = 0
for i in range(n):
    x = a + i * dx
    dy = f(x + dx) - f(x) # 풊y
    
    ds = np.sqrt(dx**2 + dy**2) # 풊s

    lengde += ds # Legger til 풊s til den totale lengden

print(f"{lengde = :.2f}")
`````

## Lengden av en parametrisert kurve

Vi skal se p친 hvordan vi kan regne ut lengden til parametriserte kurver i b친de to- og tre dimensjoner. Det viser seg at teorien er akkurat den samme for de to tilfellene, men at det i tre dimensjoner dukker opp ett ekstra ledd for $z$-retningen noen av likningene.
Uansett skal vi tenke oss at vi har parametriserte vektorer p친 formen

\begin{align*}
\vec{r}(t) & = \left[x(t), y(t)\right] \quad\quad\quad \text{(2 dimensjoner)}, \\
\\
\vec{r}(t) & = \left[x(t), y(t), z(t)\right] \quad \text{(3 dimensjoner)}. \\
\end{align*}

### Algoritme: Summere Linjeelementer $\Delta s$

Denne metoden er i stor grad uendret. Vi regner fortsatt ut lengden av hypotenusen i $n$ trekanter. Den store forskjellen er vi n친 ikke direkte setter verdiene til $x$, men lar heller parameteren $t$ bestemme hva verdien til $x$ og $y$ er. I tillegg vil den bestemme hva verdien til $z$ er n친r vi jobber i tre dimensjoner. F칮lgende resultater gir oss verkt칮yene vi trenger for 친 regne ut lengden av en parametrisert kurve:

```{prf:theorem} Lengde av kurve (2 dimensjoner)
:label: lengde_kurve_2d

La $\vec{r}(t) = \left[x(t), y(t)\right]$ v칝re en parametrisert kurve i to dimensjoner og $t \in [a, b]$ er en parameter. Vi introduserer $n + 1$ jevnt fordelte punkter i parameterrommet $t_0, t_1, t_2, \ldots, t_n$. Da er linjeelementene $\Delta s_i$ gitt ved

$$
\Delta s_i = \sqrt{\Delta x_i^2 + \Delta y_i^2},
$$

der 

\begin{align*}
\Delta x_i & = x(t_i + \Delta t) - x(t_i), \\
\Delta y_i & = y(t_i + \Delta t) - y(t_i),
\end{align*}

der $t_i = a + i\Delta t$ for $i = 0, 1, 2, \ldots, n$, og $\Delta t = (b - a) / n$. Lengden av kurven er da tiln칝rmet gitt ved

$$
L \approx \sum_{i=0}^{n-1} \Delta s_i = \Delta s_0 + \Delta s_1 + \ldots + \Delta s_{n-1}.
$$
```

```{prf:theorem} Lengde av kurve (3 dimensjoner)
:label: lengde_kurve_3d

La $\vec{r}(t) = \left[x(t), y(t), z(t)\right]$ v칝re en parametrisert kurve i tre dimensjoner og $t \in [a, b]$ er en parameter. 
La $t_0, t_1, \ldots, t_n$ v칝re $n + 1$ punkter i parameterrommet som er jevnt fordelt.
Da er linjeelementene $\Delta s_i$ gitt ved

$$
\Delta s_i = \sqrt{\Delta x_i^2 + \Delta y_i^2 + \Delta z_i^2},
$$

der 

\begin{align*}
\Delta x_i & = x(t_i + \Delta t) - x(t_i), \\
\Delta y_i & = y(t_i + \Delta t) - y(t_i), \\
\Delta z_i & = z(t_i + \Delta t) - z(t_i),
\end{align*}

der $t_i = a + i\Delta t$ for $i = 0, 1, 2, \ldots, n$ og $\Delta t = (b - a) / n$. Lengden av kurven er da tiln칝rmet gitt ved

$$
L \approx \sum_{i=0}^{n-1} \Delta s_i = \Delta s_0 + \Delta s_1 + \ldots + \Delta s_{n-1}.
$$
```

### Eksempel 4

En *heliks* er en kurve som beveger seg langs randen av en sylinder. Et eksempel p친 der denne kurven dukker opp i naturen er n친r ladde partikler fra solstormer f칮lger jordens magnetfelt inn mot polene! I {numref}`helix` ser vi en skisse av en heliks.

```{figure} ./figurer/helix.pdf
:name: helix

En tre-dimensjonal heliks. Pilhoder er tegnet inn for 친 indikere retningen til heliksen i rommet.
```

Heliksen i {numref}`helix` er beskrevet av parametriseringen

$$
\vec{r}(t) = \left[\cos(t), \sin(t), t\right], \quad t \in [0, 10\pi].
$$

Vi kan bruke metoden med linjeelementer til 친 finne lengden av heliksen p친 f칮lgende m친te:

`````python
import numpy as np

def x(t):
    return np.cos(t)


def y(t):
    return np.sin(t)


def z(t):
    return t


a = 0 # Startverdien til t
b = 10 * np.pi # Sluttverdien til t
n = 100_000 # antall linjeelementer
dt = (b - a) / n # 풊t

lengde = 0
for i in range(n):
    t = a + i * dt # Regn ut t-verdien

    # Regn ut katetene i 3-dimensjonal trekant
    dx = x(t + dt) - x(t) # 풊x
    dy = y(t + dt) - y(t) # 풊y
    dz = z(t + dt) - z(t) # 풊z

    # Regn ut hypotenusen
    ds = np.sqrt(dx**2 + dy**2 + dz**2) # 풊s

    # Legg til bidraget til lengden
    lengde += ds


print(f"{lengde = :.2f}")
`````

Deretter antar vi at vi kan bruke en konstant hastighet for 친 regne ut tiden det tar 친 g친 ett oml칮p med $t = L / v$ der $L$ er lengden, $v$ er banefarten og $t$ er perioden.     return (R + r * np.cos(3 * t) + d * np.sin(t)) * np.sin(2 * t)
