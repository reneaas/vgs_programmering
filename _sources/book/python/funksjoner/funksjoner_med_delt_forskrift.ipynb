# Funksjoner med delt forskrift

I matematikken dukker det noen ganger opp funksjoner med *delt forskrift*. Et eksempel p책 dette er 

$$
f(x) = 
\begin{cases}
x^2, & \text{hvis } x \leq 0, \\
x, & \text{hvis } x > 0.
\end{cases}
$$

I Python kan vi definere slike funksjoner ved hjelp av `if`-`else`-blokker. For eksempel kan vi definere $f$ som ovenfor slik:


`````python
def f(x):
    if x <= 0:
        return x**2
    else:
        return x 
`````

Grafisk kan vi visualere funksjonen for 책 se at den blir som forventet:

`````python
import numpy as np # Importerer numerisk bibliotek
import matplotlib.pyplot as plt # Importerer plottebibliotek

x0 = np.linspace(-2, 0, 1001)
x1 = np.linspace(0, 2, 1001)
plt.plot(x0, [f(i) for i in x0], label="$x^2$")
plt.plot(x1, [f(i) for i in x1], label="$x$")
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
plt.grid(True)
plt.legend()
plt.show()
`````

Vi kan ogs책 h책ndtere mer sammensatte delte forskrifter, som

$$
g(x) = 
\begin{cases}
x^2, & \text{hvis } x \leq 0, \\
x, & \text{hvis } 0 < x \leq 1, \\
x^3, & \text{hvis } x > 1.
\end{cases}
$$

som vi kan definere i Python med en `if`-`elif`-`else`-blokk:

`````python
def g(x):
    if x <= 0:
        return x**2
    elif 0 < x <= 1:
        return x
    else:
        return x**3
`````

`````python
import numpy as np # Importerer numerisk bibliotek
import matplotlib.pyplot as plt # Importerer plottebibliotek

x0 = np.linspace(-2, 0, 1001)
x1 = np.linspace(0, 1, 1001)
x2 = np.linspace(1, 2, 1001)
plt.plot(x0, [g(i) for i in x0], label="$x^2$")
plt.plot(x1, [g(i) for i in x1], label="$x$")
plt.plot(x2, [g(i) for i in x2], label="$x^3$")
plt.xlabel("$x$")
plt.ylabel("$g(x)$")
plt.grid(True)
plt.legend()
plt.show()
`````
